```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('cleaned.csv')
```

```python
df['Size'].plot(kind='line')
plt.show()
```

```python
df.plot(kind='scatter', x='Size', y='Rating')
```

```python
colors =['#EDAF97', '#FE938C', '#C49792', '#AD91A3', '#9D91A3']
genres = df['Genres'].value_counts().nlargest(20).plot(kind='pie', colors=colors)
genres.set_title('GENRES')
```

```python
colors =['#EDAF97', '#FE938C', '#C49792', '#AD91A3', '#9D91A3']
genres = df['Genres'].value_counts().plot(kind='barh')
genres.set_title('GENRES')
```
